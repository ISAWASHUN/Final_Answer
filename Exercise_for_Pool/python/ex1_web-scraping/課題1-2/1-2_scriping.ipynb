```python
#お店のURLを取得
import requests
from bs4 import BeautifulSoup
from time import sleep

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.action_chains import ActionChains

options = Options()
options.add_argument("--headless")  # なしでもOK
options.add_experimental_option("excludeSwitches", ["enable-logging"])
driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)
driver.maximize_window()

d_list_url = []
def selenium(page_url):
    driver.get(page_url)
    sleep(3)

    try:
        shop_url_tag = driver.find_element_by_css_selector(
            '#info-table > table > tbody > tr:nth-child(9) > td > ul > li > a'
        )
        shop_url = shop_url_tag.get_attribute("href")
        d_list_url.append({
            'URL':shop_url
        })
    except Exception as e:
        shop_url = ''
        d_list_url.append({
            'URL':shop_url
        })
        


url = 'https://r.gnavi.co.jp/area/tokyo/izakaya/rs/?sort=LOW&p={}.format()'
max_page_index = 3

for i in range(max_page_index):
    acccess_url = url.format(i+1)

    r = requests.get(acccess_url)
    r.raise_for_status()

    soup = BeautifulSoup(r.content, 'lxml')

    stores = soup.find_all("div", class_="style_restaurant__SeIVn")

    for store in stores:
        page_url = store.find("a").get("href")

        selenium(page_url)
    
df3 = pd.DataFrame(d_list_url)
df3
```

```python
#都道府県、市町村、番地
import requests
from bs4 import BeautifulSoup
from time import sleep
import re
import pandas as pd

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.action_chains import ActionChains


options = Options()
options.add_argument("--headless")  # なしでもOK
options.add_experimental_option("excludeSwitches", ["enable-logging"])
driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)
driver.maximize_window()


d_list_address = []
def selenium(page_url):
    driver.get(page_url)
    sleep(3)
    
    
    store_address_tag = driver.find_element_by_css_selector(
        '#info-table > table > tbody > tr:nth-child(3) > td > p > span.region'
    )
    store_address = store_address_tag.text
    
    
    def divide_addess(address):
            matches = re.match(r'(...??[都道府県])((?:旭川|伊達|石狩|盛岡|奥州|田村|南相馬|那須塩原|東村山|武蔵村山|羽村|十日町|上越|富山|野々市|大町|蒲郡|四日市|姫路|大和郡山|廿日市|下松|岩国|田川|大村)市|.+?郡(?:玉村|大町|.+?)[町村]|.+?市.+?区|.+?[市区町村])(.+)' , address)
            matches[1]
            matches[2]
            matches[3]
            
            d_list_address.append({
                '都道府県':matches[1],
                '市町村':matches[2],
                '番地':matches[3]
            })
            
    if __name__ == '__main__':
        address = store_address
        divide_addess(address)
   

url = 'https://r.gnavi.co.jp/area/tokyo/izakaya/rs/?sort=LOW&p={}.format()'
max_page_index = 3

for i in range(max_page_index):
    acccess_url = url.format(i+1)

    r = requests.get(acccess_url)
    r.raise_for_status()

    soup = BeautifulSoup(r.content, 'lxml')

    stores = soup.find_all("div", class_="style_restaurant__SeIVn")

    for store in stores:
        page_url = store.find("a").get("href")

        selenium(page_url)
    
df4 = pd.DataFrame(d_list_address)
df4
```
