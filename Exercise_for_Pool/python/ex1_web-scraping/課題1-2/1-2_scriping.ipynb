```python
#お店のURLを取得
import requests
from bs4 import BeautifulSoup
from time import sleep

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.action_chains import ActionChains

options = Options()
options.add_argument("--headless")  # なしでもOK
options.add_experimental_option("excludeSwitches", ["enable-logging"])
driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)
driver.maximize_window()


def selenium(page_url):
    driver.get(page_url)
    sleep(3)

    try:
        shop_url_tag = driver.find_element_by_css_selector(
            '#info-table > table > tbody > tr:nth-child(9) > td > ul > li > a'
        )
        shop_url = shop_url_tag.get_attribute("href")
        print(shop_url)
    except Exception as e:
        print('')


url = "https://r.gnavi.co.jp/area/tokyo/izakaya/rs/?sort=LOW"

r = requests.get(url)
soup = BeautifulSoup(r.text, "lxml")

stores = soup.find_all("div", class_="style_restaurant__SeIVn")

for store in stores:
    page_url = store.find("a").get("href")

    selenium(page_url)
    
```

```python
#店舗名、電話番号、メールアドレス
import requests
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd


from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.action_chains import ActionChains

options = Options()
options.add_argument("--headless")  # なしでもOK
options.add_experimental_option("excludeSwitches", ["enable-logging"])
driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)
driver.maximize_window()

d_list1 = []
def selenium(page_url):
    driver.get(page_url)
    sleep(3)
    
    store_name = driver.find_element_by_css_selector(
            '#info-name'
        ).text

    
    phone_number = driver.find_element_by_css_selector(
            '#info-phone > td > ul > li:nth-child(1) > span.number'
        ).text
    
    d_list1.append({
        '店舗名':store_name,
        '電話番号':phone_number,
        'メールアドレス':''
    })
   
    
url = 'https://r.gnavi.co.jp/area/tokyo/izakaya/rs/?sort=LOW&p={}.format()'
max_page_index = 3

for i in range(max_page_index):
    acccess_url = url.format(i+1)

    r = requests.get(acccess_url)
    r.raise_for_status()

    soup = BeautifulSoup(r.content, 'lxml')

    stores = soup.find_all("div", class_="style_restaurant__SeIVn")

    for store in stores:
        page_url = store.find("a").get("href")

        selenium(page_url)
    
df1 = pd.DataFrame(d_list1)
df1
```
